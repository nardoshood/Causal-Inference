```python
df_diagnosis = data_frame.diagnosis
Unnamed = ['Unnamed: 32', 'id' , 'diagnosis']
df = data_frame.drop(Unnamed, axis = 1)
df.head()
```

```python
ax = sns.countplot(df_diagnosis,label="Count")       # M = 212, B = 357
B, M = df_diagnosis.value_counts()
```

```python
df.describe()
data = df
```

```python
data_n_2 = (df- df.mean()) / (df.std())              # standardization
df = pd.concat([df_diagnosis,data_n_2.iloc[:,0:10]],axis=1)
df = pd.melt(df,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=df,split=True, inner="quart")
plt.xticks(rotation=90)
```

```python
# Second ten features
df = pd.concat([df_diagnosis,data_n_2.iloc[:,10:20]],axis=1)
df = pd.melt(df,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=df,split=True, inner="quart")
plt.xticks(rotation=90)
```

```python
# The rest of the features
df = pd.concat([df_diagnosis,data_n_2.iloc[:,20:31]],axis=1)
df = pd.melt(df,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=df,split=True, inner="quart")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x="features", y="value", hue="diagnosis", data=df)
plt.xticks(rotation=90)
```

```python
df = data_frame.loc[:,['radius_worst','perimeter_worst','area_worst']]
sns.set(style="white")
g = sns.PairGrid(df, diag_sharey=False)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot, lw=3)
```
